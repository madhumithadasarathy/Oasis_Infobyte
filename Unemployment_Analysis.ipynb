```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df = pd.read_csv("/content/Unemployment in India.csv")
df.head()
```

```python
#Dataframe Description
df.describe()
```

```python
#Dataframe Information
df.info()
```

```python
#Checking for null values
df.isnull().sum()
```

```python
#Dropping the null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#Shape of the dataframe
df.shape
```

```python
#Dropping the duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
#Pairplot
sns.pairplot(df)
```

```python
#Count  of employed, unemployed and labour participation
new_df = df[[' Estimated Unemployment Rate (%)',' Estimated Employed', ' Estimated Labour Participation Rate (%)']]
sns.countplot(new_df)
```

```python
#Pairplot of the Region
sns.pairplot(data = df , hue = 'Region')
```

```python
#All the regions that have unemployeds
df.Region.unique()
```

```python
#Barplot of the area
sns.barplot(df.Area)
plt.show()
```

```python
df.columns
```

```python
# Calculating average unemployment rate, employed, and labor participation rate
avg_unemployment_rate = df[' Estimated Unemployment Rate (%)'].mean()
avg_employed = df[' Estimated Employed'].mean()
avg_labor_participation_rate = df[' Estimated Labour Participation Rate (%)'].mean()
```

```python
# Calculating the maximum and minimum unemployment rates
max_unemployment_rate = df[' Estimated Unemployment Rate (%)'].max()
min_unemployment_rate = df[' Estimated Unemployment Rate (%)'].min()
```

```python
#Displaying Results
print("Analysis Results:")
print(f"Average Unemployment Rate: {avg_unemployment_rate:.2f}%")
print(f"Average Employed: {avg_employed:.2f}")
print(f"Average Labor Participation Rate: {avg_labor_participation_rate:.2f}%")
print(f"Maximum Unemployment Rate: {max_unemployment_rate:.2f}%")
print(f"Minimum Unemployment Rate: {min_unemployment_rate:.2f}%")
```

```python
# Histogram of Unemployment Rate
plt.figure(figsize=(10, 6))
sns.histplot(df[' Estimated Unemployment Rate (%)'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Line plot of Unemployment Rate over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=df, marker='o', color='orange')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Scatter plot of Unemployment Rate vs Employed
plt.figure(figsize=(10, 6))
sns.scatterplot(x=' Estimated Employed', y=' Estimated Unemployment Rate (%)', data=df, color='green')
plt.title('Unemployment Rate vs Employed')
plt.xlabel('Employed')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
# Box plot of Unemployment Rate
plt.figure(figsize=(8, 6))
sns.boxplot(df[' Estimated Unemployment Rate (%)'], color='skyblue')
plt.title('Boxplot of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

# **Results of the Analysis:**

1. **Distribution of Unemployment Rate**:
   - The histogram of the unemployment rate shows that the majority of regions have unemployment rates clustered around a certain range, with a few regions having higher or lower rates.

2. **Unemployment Rate Over Time**:
   - The line plot illustrates how the unemployment rate fluctuates over time. It can help identify trends or seasonal patterns in unemployment rates.

3. **Unemployment Rate vs Employed**:
   - The scatter plot shows the relationship between the number of employed individuals and the unemployment rate. It helps to understand if there's any correlation between these two variables.

4. **Boxplot of Unemployment Rate**:
   - The box plot provides a summary of the distribution of the unemployment rate, indicating the median, quartiles, and any outliers. It helps to identify regions with unusually high or low unemployment rates compared to the majority of regions.

Based on these visualizations and analysis, we can draw several conclusions:

- **Temporal Trends**: We can observe if there are any consistent trends or patterns in the unemployment rate over time, such as seasonal fluctuations or long-term changes.
  
- **Regional Disparities**: The analysis allows us to identify regions with higher or lower unemployment rates compared to others. This can indicate areas where targeted interventions or policies may be needed to address unemployment issues.

- **Employment Dynamics**: By examining the relationship between unemployment rate and the number of employed individuals, we can understand if there's any inverse relationship between these variables, as one would expect.

- **Outlier Detection**: The box plot helps identify regions with unusually high or low unemployment rates compared to the rest, which may require further investigation.

Overall, the analysis provides insights into the dynamics of unemployment across regions and time, facilitating informed decision-making and policy formulation to address unemployment challenges effectively.
